```python
# sin 그래프
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(0,14,100)
y = np.sin(x)
y1 = np.sin(x*0.5)
y2 = np.sin(x+1.0)
# 그림의 크기 지정
plt.figure(figsize=(10,5))
plt.plot(x,y)
plt.plot(x,y, x,y1, x,y2)
```

```python
# style 테마 
# darkgrid, whitegrid, dark, white, ticks

import seaborn as sns

sns.set_style("dark")
iris = sns.load_dataset("iris")
x = iris.petal_length.values 
sns.rugplot(x)
plt.show()
```

```python
iris
i = iris.sepal_length.values
sns.displot(i)
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips
sns.countplot(x="day",data = tips)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
titanic.head(5)
sns.countplot(x = "embark_town",data=titanic)
plt.show()
```

```python
# 요일별 식사 금액, 흡연자와 비흡연자를 비교
tips = sns.load_dataset("tips")
sns.boxplot(x="day",y="total_bill",hue="smoker",data=tips)
plt.show()
```

```python
# 요일별 식사 금액, 점심 과 저녁 비교
tips = sns.load_dataset("tips")
sns.barplot(x="day",y="total_bill",hue="time",data = tips)
plt.show() 
```

```python
# lmplot 사용
tips = sns.load_dataset("tips")
tips
sns.lmplot(x="total_bill", y= "tip",hue="smoker",data= tips, palette="Set2")
plt.show()
```

```python
# heatmap 사용
flights = sns.load_dataset("flights")
flights
fgp = flights.pivot("month","year","passengers")
plt.figure(figsize=(10,8))
sns.heatmap(fgp,cmap="YlGnBu",fmt="d")
plt.show()
```

```python
# stripplot 모든 데이터를 점으로 시각화
# jitter = True로 설정하면 위치를 무작위로 바꿔서 겹치지 않게 해줌
sns.stripplot(x="day",y="total_bill", data=tips, jitter=True)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
tit = titanic.pivot_table(index=["sex"], columns=["class"], aggfunc='size') 
sns.heatmap(tit,annot=True,fmt='d', linewidths=20)
plt.show()
```

```python
iris = sns.load_dataset("iris")
iris
sns.set_style("white")
sns.pairplot(iris,hue="species")
plt.show()
```
